```python

import pandas as pd

users_df = pd.read_csv('../01raw/users.csv')
domains_df = pd.read_csv('../01raw/domains.csv')
trainings_df = pd.read_csv('../01raw/trainings.csv')
training_assignments_df = pd.read_csv('../01raw/training_assignments.csv')
scores_df = pd.read_csv('../01raw/scores.csv')

print(users_df.isnull().sum())
print(trainings_df.isnull().sum())
```

    Cleaning User Data

```python
display(users_df)
# display(trainings_df)
# display(training_assignments_df)
```

```python
# users_df.drop_duplicates(subset=['email'], inplace=True)
users_df['role'] = users_df['role'].str.upper()
users_df
```

```python
users_df['firstName'] = users_df['firstName'].str.strip()
users_df['lastName'] = users_df['lastName'].str.strip()

users_df
```

```python
users_df = users_df.drop(columns=['gender', 'password', 'email'])
users_df.rename(columns={'id': 'userId'}, inplace=True)
users_df
```

```python
scores_df
```

    Clean Domain Data 

```python
print(domains_df.isnull().sum())
```

```python
domains_df.dropna(subset=['name'], inplace=True)
domains_df.rename(columns={'id': 'domainId'}, inplace=True)
domains_df.rename(columns={'name': 'domainName'}, inplace=True)

domains_df
```

```python
domains_df.drop_duplicates(inplace=True)
```

```python
domains_df.info()
```

```python
domains_df['domainName'] = domains_df['domainName'].str.strip()
domains_df['domainName'] = domains_df['domainName'].str.replace("_"," ")
domains_df['domainName'] = domains_df['domainName'].str.title()

domains_df
# print(domains_df.head())
```

    Clean Training Data

```python
print(trainings_df.isnull().sum())
```

```python
trainings_df.dropna(subset=['name'], inplace=True)
trainings_df
```

```python
trainings_df.drop_duplicates(inplace=True)
```

```python
trainings_df['startDate'] = pd.to_datetime(trainings_df['startDate'])
trainings_df['endDate'] = pd.to_datetime(trainings_df['endDate'])
```

```python
  # Create a new column for training duration
trainings_df['duration'] = (trainings_df['endDate'] - trainings_df['startDate']).dt.days.astype(str) + ' days'
trainings_df
```

```python
trainings_df['name'] = trainings_df['name'].str.strip()
trainings_df['name'] = trainings_df['name'].str.replace("_"," ")
trainings_df['name'] = trainings_df['name'].str.title()


trainings_df['description'] = trainings_df['description'].str.strip().replace("_"," ") 
trainings_df['description'] = trainings_df['description'].str.title()


display(trainings_df)
```

    Clean Training Assignment Data 
    

```python
# training_assignments_df.isnull().sum()
# training_assignments_df.info()

training_assignments_df.duplicated().sum()

training_assignments_df
```

    Clean Scores Data 

```python
print(scores_df.isnull().sum())
scores_df.info()
```

```python
scores_df['value'] = scores_df['value'].astype(int)
scores_df
```

```python
# Save cleaned DataFrames (optional)
users_df.to_csv('./users_cleaned.csv', index=False)
domains_df.to_csv('./domains_cleaned.csv', index=False)
trainings_df.to_csv('./trainings_cleaned.csv', index=False)
training_assignments_df.to_csv('./training_assignments_cleaned.csv', index=False)
scores_df.to_csv('./scores_cleaned.csv', index=False)
```
